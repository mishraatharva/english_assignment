```python
audio_file_path = r"U:\assignment\english_assignment\freesound_community-000981_jfk-space-race-speech-59951.mp3"
```

```python
from pydub import AudioSegment
import whisper

def generate_transcript(audio_file_path):

    # Convert mp3 to wav
    sound = AudioSegment.from_mp3(audio_file_path)
    sound.export("transcript.wav", format="wav")

    # Load Whisper model
    model = whisper.load_model("base")

    # Generate transcript
    result = model.transcribe(
        "transcript.wav",
        fp16=False
    )

    return result["text"]
```

```python
transcript = generate_transcript(audio_file_path)
```

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
Create a structured summary of the transcript.

Requirements:
1. Title
2. Executive Summary
3. Main Topics Discussed
4. Key Points (bullet points)
5. Decisions Made
6. Action Items
7. Conclusion

Transcript:
{transcript}
""")

chain = prompt | llm

summary = chain.invoke({
    "transcript": transcript
})

print(summary.content)
```

```python
from groq import Groq
import os

client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)

def generate_transcript(audio_file_path):
    with open(audio_file_path, "rb") as file:
        transcription = client.audio.transcriptions.create(
            file=file,
            model="whisper-large-v3"
        )

    return transcription.text
```
